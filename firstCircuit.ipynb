```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
```

Sur pennylane, on définit un circuit à l'aide de deux éléments :
- un device
- une fonction quantique

```python
dev = qml.device("default.qubit", wires=2)
```

```python
@qml.qnode(dev)
def circuit(theta):
    qml.PauliX(wires=1)
    qml.CNOT(wires=[1, 0])
    qml.RY(theta, wires=0)
    return qml.expval(qml.PauliZ(0))

```

```python
print(circuit(np.pi))
```

```python
thetas = np.arange(-np.pi, np.pi, 0.01)
measuremens = np.zeros(len(thetas))

for i, theta in enumerate(thetas):
    measuremens[i] = circuit(theta)

fig, ax = plt.subplots()
ax.plot(measuremens)
```
